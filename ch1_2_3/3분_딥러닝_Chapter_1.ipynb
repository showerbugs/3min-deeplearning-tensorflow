# Chapter 1 딥러닝과 텐서플로의 만남

## 1.1 인공지능, 머신러닝 그리고 딥러닝

`인공지능`은 관념에 가까운 것으로, 컴퓨터가 인간의 사고를 모방하는 모든 것을 뜻한다.

`머신러닝`은 개념으로써, 컴퓨터가 주어진 데이터를 이용해 스스로 학습하는 것을 말합니다.

`딥러닝`으로 대표되는 `인공신경망`은 머신러닝을 구현하는 기술의 하나로, 인간 뇌의 동작 방식에서 착안하여 개발한 학습 방법입니다.

![image/1_deeplearning.png](image/1_deeplearning.png)

예를 들어 기존의 rule-based AI가 바나나를 인식하려면 인간이 컴퓨터에 특징(feature)들을 알려줘야 했습니다.

하지만 머신러닝의 학습 방식은 바나나 사진만 주면 컴퓨터가 스스로 바나나의 특징을 학습합니다.

![image/1_banana.png](image/1_banana.png)

만약 AI가 썰어놓은 바나나를 추가로 인식하려면

- rule-based AI는 하얀색이고, 납작하고, 둥글다를 수동으로 추가해야 함
- machine learning은 썰어놓은 바나나 사진만 입력해도 알아서 특징을 학습하고 인식 할 수 있음

![image/1_banana2.png](image/1_banana2.png)

풀고자 하는 문제가 복잡할 수록 rule-based AI 보다는 Machine Learning을 이용하는 것이 효율적으로 해결할 수 있어집니다.

> 수작업으로 특징을 넣어서 알고리즘을 짜는 것보다 컴퓨터가 특징을 추출하는 쪽이 더 빠를 경우에

인간도 파악하기 어려울 정도로 복잡한 문제는 머신러닝도 한계가 있었는데,, 딥러닝이 등장하면서 발전 속도가 급격해졌다.

2012년 ILSVRC 대회에서 neural network를 이용한 AlexNet이 나와 75% -> 84.7%로 인식률을 확 높였습니다.

- 이후 Deep Learning 이 상위 랭크를 휩쓸었고
- 지금은 인간의 인식률(95%)을 뛰어넘는 97%에 육박

![error_rate](image/1_error_rate.png)

딥러닝은 1940년부터 연구되었는데 지금에서야 뜨게 된거는

1. GPU 발전
    - 병렬처리 능력
2. 다양한 딥러닝 알고리즘 발명
    - 역전파 알고리즘
3. 인터넷의 발전으로 데이터 수급 문제 해결
    - 데이터에서 feature를 발견하기 때문에 그 양이 방대할 수록 정확도가 높아짐
4. 온라인 강의, 오픈 소스
